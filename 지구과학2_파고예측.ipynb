# 결측치 제거


```python
data = data.dropna() # 결측치 제거

data.isnull().sum() # 확인과정
```

# 데이터 전처리



```python
n_data= data['평균 유의 파고(m)']

y_data = [1 if height > 0.3 else 0 for height in n_data]

print(y_data)
```

```python
x_data=[]

for i, rows in data.iterrows():
    x_data.append([
        rows['최고 수온(°C)'],
        rows['평균 수온(°C)'],
        rows['최저 수온(°C)'],
        rows['평균 파주기(sec)'],
        rows['최고 파주기(sec)'],
        rows['평균풍속(m/s)'],
        rows['최대풍속(m/s)'],
        rows['최대풍속풍향(deg)'],
        rows['최대풍속시각'],
        rows['최대순간풍속(m/s)'],
        rows['평균 파고(m)'],
        rows['최고 유의 파고(m)']
        # rows['최대순간풍속풍향(deg)']
    ])




print(x_data)
    # print(rows['평균 수온(°C)'])
    # print(rows['최고 수온(°C)'])
    # print(rows['최저 수온(°C)'])
    # print(rows['평균 파주기(sec)'])
    # print(rows['최고 파주기(sec)'])
    # print(rows['평균풍속(m/s)'])
    # print(rows['최대풍속(m/s)'])
    # print(rows['최대풍속풍향(deg)'])
    # print(rows['최대풍속시각'])
    # print(rows['최대순간풍속(m/s)'])
    # print(rows['최대순간풍속풍향(deg)'])
```

# 데이터 상관관계 파악

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# x_data를 DataFrame으로 변환
x_data_df = pd.DataFrame(x_data, columns=[
    'Highest Temperature(°C)',
    'Average Temperature(°C)',
    'Lowest Temperature(°C)',
    'Average Wave Period(sec)',
    'Highest Wave Period(sec)',
    'Average Wind Speed(m/s)',
    'Maximum Wind Speed(m/s)',
    'Max Wind Speed Direction(deg)',
    'Max Wind Speed Time',
    'Max Gust Speed(m/s)',
    'Average Wave Height(m)',
    'Highest Significant Wave Height(m)'
])

# n_data를 DataFrame으로 변환
n_data_df = pd.DataFrame(n_data)

# 데이터 결합
combined_data = pd.concat([x_data_df, n_data_df], axis=1)

# 상관 관계 매트릭스 계산
correlation_matrix = combined_data.corr()

# 상관 관계 매트릭스 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix between Features and Average Significant Wave Height(m)')
plt.show()

```

# **모델 구성 및 학습**

```python
# 학습 드가자~~~~~~~~~~~~~~~~~~~~~~~

import tensorflow as tf
import numpy as np
import pandas as pd

# 넘파이 변환
x_data = np.array(x_data)
n_data = np.array(n_data)

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='tanh'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(tf.keras.optimizers.RMSprop(learning_rate=0.001), loss='mse', metrics=['accuracy'])

model.fit( x_data, n_data, epochs=100)
```

**시각화**

```python
import matplotlib.pyplot as plt

history = model.fit(x_data, y_data, epochs=1000)

# Plot training accuracy

print('이건 정확도 그래프이구요')

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training Accuracy Over Epochs')
plt.ylim(0, 1)
plt.legend()
plt.show()

print('아래건 손실함수입니다')

plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
